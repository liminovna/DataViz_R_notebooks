```r
# Tidyverse and ggplot
```

```r
library(tidyverse)
```

```r
df <- read_csv("maltese_ft.csv", na = c("-", "-99"))
```

```r
top_n(df,10)
```

```r
options(repr.plot.width = 80, repr.plot.height = 7, repr.plot.res = 100)
```

```r
ggplot(
  aes(x=Word, y=Occurrences, size=500, color = Gender, shape = Number),
  data=df
) + theme_bw() + geom_point(alpha=.75) + xlab("Maltese nouns") + ylab("Absoute frequency") + ggtitle("Maltese noun frequency distribution") + theme(plot.title = element_text(hjust=0.5))
```

```r
plot <- ggplot(
  aes(x=Word, y=Occurrences, size=500, color = Gender, shape = Number),
  data=df
) +
  theme_bw() +
    geom_point(alpha=.75) +
      geom_text(aes(label=Word), check_overlap = TRUE, vjust=2.5) +
      xlab("Maltese nouns") + ylab("Absoute frequency") +
        ggtitle("Maltese noun frequency distribution") +
          scale_y_log10(guide = "axis_logticks") +
            theme(axis.ticks.x = element_blank(), axis.text.x = element_blank())

plot
```

```r
ggsave(plot, file="plot2.png",dpi=300)
```
