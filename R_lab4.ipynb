```r
library(tibble)
library(tidyr)
library(tidyverse)
```

```r
tibble(x = 1:100, PDF = dnorm(x = x, mean = 20, sd = 5))  %>%
  ggplot(aes(x, PDF))+
  geom_point()+
  geom_line()+
  labs(title = "График нормального распределения (μ = 50, sd = 5)")
```

```r
tibble(x = 1:100,
       CDF = pnorm(x, mean = 50, sd = 10)) %>%
  ggplot(aes(x, CDF))+
  geom_point()+
  geom_line()+
  labs(title = "CDF нормального распределения (μ = 50, sd = 10)")
tibble(quantiles = seq(0, 1, by = 0.01),
       value = qnorm(quantiles, mean = 50, sd = 10)) %>%
  ggplot(aes(quantiles, value))+
  geom_point()+
  geom_line()+
  labs(title = "inverse CDF нормального распределения (μ = 50, sd = 10)")
tibble(sample = rnorm(100000, mean = 50, sd = 10)) %>%
  ggplot(aes(sample))+
  geom_histogram()+
  labs(title = "выборка нормально распределенных чисел (μ = 50, sd = 10)")
```

```r
vowels <- read_csv("https://raw.githubusercontent.com/agricolamz/2021_da4l/master/data/phonTools_hillenbrand_1995.csv")
summary(vowels)
```

```r
vowels %>%
  ggplot(aes(dur)) +
  geom_histogram(aes(y = after_stat(density))) + # обратите внимание на аргумент after_stat(density)
  stat_function(fun = dnorm, args = list(mean = 274.673, sd = 64.482), color = "red")
```

```r
diet <- read_csv('https://raw.githubusercontent.com/Pozdniakov/tidy_stats/master/data/stcp-Rdataset-Diet.csv')
diet
```

```r
diet_y <- diet %>%
  filter(Diet == 1)
```

```r
t.test(diet$pre.weight, diet$weight6weeks, paired=TRUE)
# p-value меньше критического значения
```

```r
# 2.2e-16 -- самое маленькое значение близкое к 0
```

```r
t.test(diet_y$pre.weight, diet_y$weight6weeks, paired=TRUE)
```

```r
diet_long <- diet %>% pivot_longer(
  cols=c(pre.weight, weight6weeks),
  names_to = "when",
  values_to = "weight"
)


t.test(weight ~ when, data = diet_long, paired=TRUE)
```
